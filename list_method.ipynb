#### Append
** Adds an item to the very end of the list **
- append(element) 

```python
h1=['a','b','c','d']
h1.append('e')
print(f"h1 element append={h1}")

h2=['Ankita','Anvika','Neha']
h2.append('Pranjal')
print(f"h2 element append={h2}")

h3=[1,2,3,4,5]
h3.append(6)
print(f"h3 element append={h3}")
```

#### insert(index, element): 
** Squeezes an item into a specific position **

```python
h1=['a','b','c','d']
h1.insert(0,'g')
print(f"new element insert={h1}")

h2=['Ankita','Anvika','Neha']
h2.insert(3,'Priya')
print(f"new element insert={h2}")

h3=[1,2,3,4,5]
h3.insert(5, 6)
print(f"new element insert={h3}")
```

### extend(iterable): 
** Appends elements from another list (or any iterable) to the current one. **

```python
h1=['a','b','c','d']
h2=['e','f']
h1.extend(h2)
print(f" add element fron another list={h1}")

h3=['Ankita','Neha','Megha']
h4=['Pranjal','Nita']
h3.extend(h4)
print(f"add element from another list={h3}")

h5=[1,2,3,4]
h6=[5,6]
h5.extend(h6)
print(f"add element from another list={h5}")
```

### pop(index): 
** Removes and returns the item at the given index (defaults to the last item).**

```python
h1=['a','b','c','d']
h1.pop(1)
print(f"return element={h1}")

h2=['Ankita','Anvika','Neha']
h2.pop(0)
print(f"return element={h2}")

h3=[1,2,3,4,5]
h3.pop(3)
print(f"return element={h3}")
```

### remove(value): 
** Finds and deletes the first occurrence of a specific value.**

```python
h1=['a','b','c','d']
h1.remove('c')
print(f"return element={h1}")

h2=['Ankita','Anvika','Neha']
h2.remove('Ankita')
print(f"return element={h2}")

h3=[1,2,3,4,5]
h3.remove(3)
print(f"return element={h3}")
```

### sort(key=..., reverse=...): 
** Sorts the list in place (ascending by default) **

```python
h1=['b','p','f','a']
h1.sort()
print(f" Asending order ={h1}")

h2=['b','p','f','a']
h2.sort(reverse=True)
print(f" desending order={h2}")

h3=['Ankita','Anvika','Neha']
h3.sort(key=len)
print(f" according to len={h3}")

```

### reverse(): 
** Flips the order of the list items **

```python
h1=['a','b','c','d']
h1.reverse()
print(f"reverse order={h1}")

h2=['Ankita','Anvika','Neha']
h2.reverse()
print(f"reverse ordert={h2}")

h3=[1,2,3,4,5]
h3.reverse()
print(f"reverse ordert={h3}")
```

### index(value): 
** Returns the index of the first element with the specified value.**

```python
h1=['a','b','c','d']
print(f" return index={h1.index('c')}")


h2=['Ankita','Anvika','Neha']
print(f"return index={h2.index('Neha')}")

h3=[1,2,3,4,5]
print(f"return index={h3.index(4)}")
```

### count(value): 
** Returns the number of times a value appears in the list.**

```python
h1=['a','b','c','d','a']
print(f" count a={h1.count('a')}")


h2=['Ankita','Anvika','Neha','Anvika']
print(f"count Anvika={h2.count('Anvika')}")

h3=[1,2,3,4,2,5,2]
print(f"count number 2={h3.count(2)}")
```

### clear
** Nukes the entire list, leaving it empty.**

```python
h1 = ['a', 'b', 'c', 'd', 'a']
h1.clear()
print(f"h1 = {h1}")

h2 = ['Ankita', 'Anvika', 'Neha', 'Anvika']
h2.clear() 
print(f"h2 = {h2}")

h3 = [1, 2, 3, 4, 2, 5, 2]
h3.clear()
print(f"h3 = {h3}")
```
